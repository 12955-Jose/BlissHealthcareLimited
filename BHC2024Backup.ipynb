```python

import pandas as pd
from typing import Union
import calendar
from datetime import datetime

#LOAD BUDGET TABLES

def load_Budget(file_path):
    # Read the data from CSV file with appropriate encoding and column data types
    df = pd.read_csv(file_path, encoding='utf-8', thousands=',', dtype={
        'Month': str,
        'Value': float,
        'Report': str,
        'Region': str,
        'Location_Name': str,
        'Sceme': str
    })

    df['Month'] = pd.to_datetime(df['Month'])


    return df

# Load Budget data using the cached function
Allbudget_df = load_Budget('c:/ProgramData/MySQL/MySQL Server 8.0/Uploads/Streamlit/Budget_Final_2023.csv')

# Filter Allbudget_df for Revenue
revenue_budget_df = Allbudget_df[Allbudget_df['Report'] == 'Revenue']
# Group by 'Month' and 'Scheme' and calculate sum of 'Value' for each group
ReveneuBudgetsummary_df = revenue_budget_df.groupby(['Month', 'Sceme','Location_Name','Region'], as_index=False,observed=False)['Value'].sum()

# Filter Allbudget_df for Footfall
footfall_budget_df = Allbudget_df[Allbudget_df['Report'] == 'Footfalls']
# Group by 'Month' and 'Scheme' and calculate sum of 'Value' for each group
FootfallBudgetsummary_df = footfall_budget_df.groupby(['Month', 'Sceme','Location_Name','Region'], as_index=False,observed=False)['Value'].sum()

# Rename the 'Sceme' column to 'Scheme' in both dataframes
ReveneuBudgetsummary_df = ReveneuBudgetsummary_df.rename(columns={'Sceme': 'Scheme','Location_Name':'location_name' })
FootfallBudgetsummary_df = FootfallBudgetsummary_df.rename(columns={'Sceme': 'Scheme','Location_Name':'location_name'})


# Merge monthly_scheme_summary_df with FootfallBudgetsummary_df and ReveneuBudgetsummary_df
Budgetmerged_df = pd.merge(ReveneuBudgetsummary_df, FootfallBudgetsummary_df, how='left', on=['Month', 'Scheme','location_name','Region'])


# Rename the columns in the merged DataFrame
Budgetmerged_df = Budgetmerged_df.rename(columns={
    'Value_x': 'Budget_Revenue',
    'Value_y': 'Budget_Footfalls'
})

# Convert 'Month' to datetime
Budgetmerged_df['bill_date'] = pd.to_datetime(Budgetmerged_df['Month'], format='%B', yearfirst=True)

# Create 'bill_date' column with all dates for each day in the month
Budgetmerged_df['bill_date'] = Budgetmerged_df.apply(lambda row: pd.date_range(row['bill_date'], row['bill_date'] + pd.offsets.MonthEnd(0), freq='D'), axis=1)

# Explode the 'bill_date' column to separate rows for each date
Budgetmerged_df = Budgetmerged_df.explode('bill_date')

Budgetmerged_df['Month'] = pd.to_datetime(Budgetmerged_df['Month'])
Budgetmerged_df['Month'] =Budgetmerged_df['Month'].dt.month.apply(lambda x: calendar.month_name[x])

# Calculate the number of days in each month and divide 'Budget Revenue' by that number
Budgetmerged_df['days_in_month'] = Budgetmerged_df['bill_date'].dt.daysinmonth

#Calculate the Revenue per day
Budgetmerged_df['Budget_Revenue'] = Budgetmerged_df['Budget_Revenue']/Budgetmerged_df['days_in_month']

#Calculate the fooftafalls per day
Budgetmerged_df['Budget_Footfalls'] = Budgetmerged_df['Budget_Footfalls']/ Budgetmerged_df['days_in_month']

# Drop the 'days_in_month' column if you don't need it anymore
Budgetmerged_df = Budgetmerged_df.drop(columns=['days_in_month'])

# Calculate total budget footfalls and total budget revenue
total_budget_footfalls = Budgetmerged_df['Budget_Footfalls'].sum()
total_budget_revenue = Budgetmerged_df['Budget_Revenue'].sum()

print(f'Total Budget Footfalls: {total_budget_footfalls}')
print(f'Total Budget Revenue: {total_budget_revenue}')

# Convert 'bill_date' column in Budgetmerged_df to datetime
Allschemes_daily['bill_date'] = pd.to_datetime(Allschemes_daily['bill_date'])

# Merge Allschemes_daily with Budgetmerged_df
Allmerged_df = pd.merge(Budgetmerged_df,Allschemes_daily, how='left', on=['bill_date', 'Scheme', 'location_name'])

# Convert 'bill_date' column in Allmerged_df to datetime
Allmerged_df['Year'] = pd.to_datetime(Allmerged_df['bill_date']).dt.year

# Export to CSV
Allmerged_df.to_csv('c:/ProgramData/MySQL/MySQL Server 8.0/Uploads/Streamlit/Allmerged.csv', index=False)

Alldepartment_daily.to_csv('c:/ProgramData/MySQL/MySQL Server 8.0/Uploads/Streamlit/Alldepartment.csv', index=False)

Allmerged_df = Allmerged_df.round(0)
```
